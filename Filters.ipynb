```python
import numpy as np
import matplotlib.pyplot as plt
import time 
%matplotlib inline
```

## Creating Data

```python
fs = 2000 #Hz
t = np.linspace(0, 2, 2*fs) #seconds, time
l_freq = np.sin(10*2*np.pi*t) #10 Hz
h_freq = np.sin(500*2*np.pi*t) #500 Hz
nt = np.sin(60*2*np.pi*t) #60 Hz
signal = l_freq + h_freq + nt

#Plotting 
plt.title("Time Domain")
plt.plot(t, signal)
plt.xlabel("Time(s)")
plt.show()

plt.title("Frequency Domain")
f = np.linspace(0, t.shape[0]/2., t.shape[0])
plt.plot(f,np.abs(np.fft.fft(signal)))
plt.xlabel("Frequency (Hz)")
plt.xlim(0, fs/2) #showing  only the useful part
plt.show()
```

## Filters (in frequency domain)

```python
def HP(signal, fc, fs):
    s_kernel = int(round(signal.shape[0]*(fc/fs)))
    kernel = np.hstack((np.zeros(s_kernel),np.ones(signal.shape[0]- 2*s_kernel), np.zeros(s_kernel)))
    Y = np.multiply(kernel, np.fft.fft(signal))
    return np.fft.ifft(Y).real


def LP(signal, fc, fs):
    s_kernel = int(round(signal.shape[0]*(fc/fs)))
    kernel = np.hstack((np.ones(s_kernel), np.zeros(signal.shape[0]- 2*s_kernel), np.ones(s_kernel)))
    Y = np.multiply(kernel, np.fft.fft(signal))
    return np.fft.ifft(Y).real


def notch(signal, fc, q, fs):
    s_kernel = int(round(signal.shape[0]*(fc/fs)))
    kernel = np.hstack((np.ones(s_kernel - int(q/2)), np.zeros(q), np.ones(signal.shape[0]-q - 2*s_kernel), np.zeros(q),np.ones(s_kernel - int(q/2))))
    Y = np.multiply(kernel, np.fft.fft(signal))
    return np.fft.ifft(Y).real

```

## Testing and timing

### Low pass

```python
start = time.time()
lp_signal  = LP(signal, 450., fs)
end = time.time()

print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"

#Plotting 
plt.title("New time Domain")
plt.plot(t, lp_signal)
plt.xlabel("Time(s)")
plt.show()

plt.title("New frequency Domain")
plt.plot(f,np.abs(np.fft.fft(lp_signal)))
plt.xlabel("Frequency (Hz)")
plt.xlim(0, fs/2) #showing  only the useful part
plt.show()
```

### High pass

```python
start = time.time()
hp_signal  = HP(signal, 15., fs)
end = time.time()

print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"

#Plotting 
plt.title("New time Domain")
plt.plot(t, hp_signal)
plt.xlabel("Time(s)")
plt.show()

plt.title("New frequency Domain")
plt.plot(f,np.abs(np.fft.fft(hp_signal)))
plt.xlabel("Frequency (Hz)")
plt.xlim(0, fs/2) #showing  only the useful part
plt.show()
```

### Notch filter

```python
start = time.time()
no_signal  = notch(signal, 60., 10, fs)
end = time.time()

print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"

#Plotting 
plt.title("New time Domain")
plt.plot(t, no_signal)
plt.xlabel("Time(s)")
plt.show()

plt.title("New frequency Domain")
plt.plot(f,np.abs(np.fft.fft(no_signal)))
plt.xlabel("Frequency (Hz)")
plt.xlim(0, fs/2) #showing  only the useful part
plt.show()
```

### All filters

```python
start = time.time()
all_signal  = HP(LP(notch(signal, 60., 10, fs), 450, fs), 15, fs)
end = time.time()

print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"

#Plotting 
plt.title("New time Domain")
plt.plot(t, all_signal)
plt.xlabel("Time(s)")
plt.show()

plt.title("New frequency Domain")
plt.plot(f,np.abs(np.fft.fft(all_signal)))
plt.xlabel("Frequency (Hz)")
plt.xlim(0, fs/2) #showing  only the useful part
plt.show()
```

## Filters by convolution

```python
#These functions return the sinc(t), convolution must be used
def HP_conv(signal_size, fc, fs):
    s_kernel = int(round(signal_size*(fc/fs)))
    kernel = np.hstack((np.zeros(s_kernel),np.ones(signal_size- 2*s_kernel), np.zeros(s_kernel)))
    return np.fft.fftshift(np.fft.ifft(kernel).real)


def LP_conv(signal_size, fc, fs):
    s_kernel = int(round(signal.shape[0]*(fc/fs)))
    kernel = np.hstack((np.ones(s_kernel), np.zeros(signal_size- 2*s_kernel), np.ones(s_kernel)))
    return np.fft.fftshift(np.fft.ifft(kernel).real)


def notch_conv(signal_size, fc, q, fs):
    s_kernel = int(round(signal.shape[0]*(fc/fs)))
    kernel = np.hstack((np.ones(s_kernel - int(q/2)), np.zeros(q), np.ones(signal_size-q - 2*s_kernel), np.zeros(q),np.ones(s_kernel - int(q/2))))
    return np.fft.fftshift(np.fft.ifft(kernel).real)
```

## Testing and timing

### Visualizing  the sinc(t) function

```python
start = time.time()
sinc_hp = HP_conv(signal.shape[0], 15., fs)
end = time.time()
plt.title("High pass")
plt.plot(sinc_hp)
plt.show()
print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"


start = time.time()
sinc_lp = LP_conv(signal.shape[0], 450., fs)
end = time.time()
plt.title("Low pass")
plt.plot(sinc_lp)
plt.show()
print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"


start = time.time()
sinc_no = notch_conv(signal.shape[0], 60.,20, fs)
end = time.time()
plt.title("Notch")
plt.plot(sinc_no)
plt.show()
print "Processing time: ", end-start, "seconds"
print "Processing time per sample: ", (end-start)/signal.shape[0], "seconds"
```

### High pass by conv

```python
start = time.time()
hp_signal  = np.convolve(HP_conv(signal.shape[0], 15., fs), signal, 'same')
end = time.time()
print "Processing time: ", end-start, "seconds"

#Partial time
start = time.time()
sinc = HP_conv(signal.shape[0], 15., fs)
end = time.time()
p1 = end - start
start = time.time()
hp_signal  = np.convolve(sinc, signal, 'same')
end = time.time()
p2 = end - start
print "Partial 1:", p1, "Partial 2:", p2
print "Total:", p1+p2

#Plotting 
plt.title("New time Domain")
plt.plot(t, hp_signal)
plt.xlabel("Time(s)")
plt.show()

plt.title("New frequency Domain")
plt.plot(f,np.abs(np.fft.fft(hp_signal)))
plt.xlabel("Frequency (Hz)")
plt.xlim(0, fs/2) #showing  only the useful part
plt.show()
```
